```python

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df['Total_Venta'], kde=True)
plt.title('Distribuci贸n de la Venta Total')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Total_Venta'])
plt.title('Boxplot de la Venta Total')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(y=df['Categoria_Producto'], order=df['Categoria_Producto'].value_counts().index)
plt.title('Conteo de Ventas por Categor铆a de Producto')
plt.show()
```

```python
columnas_numericas = df.select_dtypes(include=['number']).columns
sns.heatmap(df[columnas_numericas].corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlaci贸n')
plt.show()

sns.scatterplot(x='Precio_Unitario', y='Total_Venta', data=df)
plt.title('Precio Unitario vs. Venta Total')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Categoria_Producto', y='Total_Venta', data=df)
plt.title('Venta Total por Categor铆a de Producto')
plt.show()
```
